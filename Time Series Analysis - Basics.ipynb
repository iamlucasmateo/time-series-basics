```python
# time series we will use
PG = wb.DataReader('PG',data_source='yahoo',start='1995-1-1')
# let's exclude the last COVID-19 shock
PG = PG[PG.index < '2020-01-31']
PG.head(5)
```

### Rolling Mean calculations

pandas' rolling window function creates a moving array over which we can do rolling calculations (such as mean, standard deviation). The first argument is the window period (the longer, the smoother the resulting curve). The other argument, min_periods, regards the initial values (which do not have a previous window). Setting it to 1 means that the array will initially be of length 1, then 2, etc, until the window size is reached. Let's see how it works.

```python
# 250 is the number of yearly working days
rolling_window = PG['Close'].rolling(250,min_periods = 1)
rolling_window.mean().to_frame().head()
```

What is pandas doing? It's taking the mean until the selected index. Like this: 

```python
for i in range(5):
    print(f'Mean {i}: {PG["Close"][0:i+1].mean()}')
```

If we move forward, as soon as we have enough data the window period will kick in. For example:

```python
rolling_window.mean().to_frame()[500:505]
```

```python
for i in range(5):
    print(f'Mean {500+i}: {PG["Close"][501+i-250:501+i].mean()}')
```

```python
# Let's plot
plt.figure(figsize=(17,4))
plt.plot(PG['Close'])
plt.plot(PG['Close'].rolling(250,min_periods=1).mean(),c='red')
plt.show()
```

```python
# Smoother: change window size
plt.figure(figsize=(17,4))
plt.plot(PG['Close'])
plt.plot(PG['Close'].rolling(2000,min_periods=1).mean(),c='red')
plt.show()
```

### Exponential Smoothing

```python
len(PG['Close'])
```

```python
def exp_smooth(series,alpha):
    '''
    series - list, array, pd.Series
    alpha - smoothing parameters
    returns smoothed array / list
    '''
    prediction = series[0]
    smoothed = [prediction]
    for i in range(1,len(series)):
        prediction = series[i]*alpha + prediction*(1-alpha)
        smoothed.append(prediction)
    return smoothed
```

```python
smoothed = exp_smooth(PG['Close'],0.005)
smoothed[0:7]
```

```python
plt.figure(figsize=(17,4))
plt.plot(PG['Close'].values)
smoothed = exp_smooth(PG['Close'],0.05)
plt.plot(smoothed,c='red')
plt.show()
```

```python
# Smoother: reduce alpha size
plt.figure(figsize=(17,4))
plt.plot(PG['Close'].values)
smoothed = exp_smooth(PG['Close'],0.005)
plt.plot(smoothed,c='red')
plt.show()
```

### Double Exponential Smoothing

```python
def double_exp_smooth(series,alpha,beta):
    '''
    series - list, array, pandas Series 
    alpha - smoothing parameter for trend  
    beta - smoothing parameter for intercept
    
    returns double smoothed series as a list
    '''
    intercept, trend = series[0], series[1] - series[0]
    smoothed = [series[0],series[1]]
    for i in range(2,len(series)):
        intercept_lag, trend_lag = intercept, trend
        intercept = alpha * series[i] + ( 1 - alpha ) * ( trend_lag + intercept_lag )
        trend = beta * ( trend - trend_lag ) + ( 1 - beta )*( trend_lag )
        smoothed.append( trend + intercept )
    return smoothed
```

```python
smoothed = double_exp_smooth(PG['Close'],0.005,0.005)
smoothed[0:7]
```

```python
plt.figure(figsize=(17,4))
plt.plot(PG['Close'].values)
smoothed = double_exp_smooth(PG['Close'],0.005,0.01)
plt.plot(smoothed,c='red')
plt.show()
```
